```python
def mathsbpf(x0, x1, n):
    a = slope(1, bounds=(x0, x1))
    f = a
    t = slope(1)
    dx = (x1 - x0) / (n-1)
    f = (t**2).sin()[x0:x1:dx].integrated()
    return f.map(n)

def fresnelnp(x0, x1, n):
    t = np.linspace(x0, x1, n)
    f = np.sin((t**2))
    
    
```

```python
from math import *

def f():
    # theta = linear(0, 0, 2*pi, 2*pi)
    theta = slope(1, bounds=(0, 2*pi))
    #r = 2.5 + (400*theta).cos() + (403*theta).cos()*0.5 + (410*theta).cos()*0.3
    r0 = (400*theta).cos()
    r1 = (403*theta).cos()*0.5
    r2 = (410*theta).cos()*0.3
    r = sum_(r0, r1, r2)
    return r.map(r.dxton(0.001))
    
```
