```python
cfg = CoreConfig(quantComplexity='low')
ch = Chain(r"""
4D:1/5
4E:1/5
4F:1/10+1/6
4C:0
4E-:0
4D:1/4
4E:1/6
4F:1/4

""")
qs = ch.quantizedScore(config=cfg)
qs.show()
```

```python
p = qs.parts[0]
p.addSpanner('beam', 0, 1-F(1, 1000))
p.measures[0].setBeamSubdivisions(0, F(1, 8))
qs.show()
```
