```python
from maelzel.core import *
```

```python
ch = Chain([Note(i, F(1,15)) for i in range(60, 75)])
ch
```

```python
ch = Chain([Note(i, F(1,15)) for i in range(60, 75)])
ch[0].addSymbol(symbols.QuantHint((15,)))
ch
```

```python
ch = Chain([Note(i, F(1,15)) for i in range(60, 75)])
ch[0].addSymbol(symbols.QuantHint((3,)*5, strength=1000000))
ch
```

```python
ch = Chain([Note(i, F(1,12)) for i in range(60, 84)])
ch
```

```python
ch = Chain([Note(i, F(1,12)) for i in range(60, 84)])
ch[0].addSymbol(symbols.QuantHint((4, 4, 4)))
ch
```

```python
ch = Chain([Note(i, F(1,12)) for i in range(60, 84)])
ch.addSymbolAt(symbols.QuantHint((6, 6)), 0)
ch.addSymbolAt(symbols.QuantHint((6, 6), strength=1000000), 1)
ch
```

It is possible to force any quantization with enough strength

```python
ch = Chain([Note(i, F(1,12)) for i in range(60, 84)])
ch.addSymbolAt(symbols.QuantHint((3, 8), strength=1000000), 0)
ch.addSymbolAt(symbols.QuantHint((7, 6), strength=1000000), 1)
ch
```
