```python
from maelzel.core import *
```

```python
scale = Chain(Note(n, 1.5) for n in ["4C", "4D+", "4Eb", "4F+", "4G"])
group = Group([Note(n, 5, start=i*2) for i, n in enumerate(["4C", "4D+", "4Eb", "4F-", "4G"])])
scale
```

```python
group.show()
group.play(instr='piano', position=0)
```

```python
evs = (
    group.events(instr='sin', gain=0.02, fade=(0.4, 2.5), position=0.5, fadeshape='cos') + 
    group.events(instr='piano', gain=0.4))
play.playEvents(evs)
```

```python
with play.OfflineRenderer() as r:
    scale.play(instr='piano', position=0.15)
```

```python
sample = r.renderedSample()
sample
```
