```python
from maelzel.core import *
```

```python
v1 = Voice(r'''
3D:1/4
3Ab:1/4
4D:1/4
4G+:1/4
4A:1/4
5F-:1/4
2D:1/4
4Gb:1/4
4Bb:1/4

''', name='v1', maxStaves=2)

v1
```

```python
from maelzel.scoring import clefutils
```

```python
ns = v1.scoringEvents()
ns
```

```python
clefutils.bestClefCombination(ns, maxStaves=2)
```
