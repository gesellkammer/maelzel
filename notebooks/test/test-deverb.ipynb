```python
from maelzel.snd import deverb
from maelzel.snd.audiosample import Sample
import sndfileio
```

```python
samples, info = sndfileio.sndget("../snd/piano-pitchtest-Cmaj.flac")
info
```

```python
s = Sample(samples, sr=info.samplerate)
s
```

```python
outsamples = deverb.removeSustain(samples=samples, sr=info.samplerate, verbose=False, morphTime=0.05)
s = Sample(outsamples, sr=info.samplerate)
s
```
