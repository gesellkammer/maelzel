```python
from maelzel.snd import deverb
from maelzel.snd.audiosample import Sample
import sndfileio
```

```python
samples, info = sndfileio.sndget("../snd/piano-pitchtest-Cmaj.flac")
info
```

```python
s = Sample(samples, sr=info.samplerate)
s
```

```python
outsamples = deverb.removeSustain(samples=samples, sr=info.samplerate, verbose=False, morphTime=0.05, transientMargin=0.1, fftsize=1024, wet=0.9)
s = Sample(outsamples, sr=info.samplerate)
s
```

```python
samples2, info2 = sndfileio.sndget("../snd/oking2.ogg")
outsamples = deverb.removeSustain(samples=samples2, sr=info2.samplerate, verbose=False, morphTime=0.05)
s = Sample(outsamples, sr=info.samplerate)
s
```
