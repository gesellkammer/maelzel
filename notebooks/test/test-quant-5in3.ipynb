```python
from maelzel.core import *
import maelzel.scoring.common
import maelzel.scoring.quant

maelzel.scoring.common.logger.setLevel("DEBUG")
maelzel.scoring.quant.logger.setLevel("DEBUG")
w = ws()
w.config['quant.debug'] = True
w.config['quant.complexity'] = 'exact'
w.scorestruct = """
2/4, q=54
"""

dur = F(1,4)*F(4,7)*F(6,7)
ch = Chain([
    "4C:1/4*4/7",
    Note("4D", dur),
    Note("4E", dur),
    Note("4F", dur),
    Note("4G", dur),
    Note("4A", dur),
    Note("4B", dur),
    Note("5C", dur),
    
])

print(w.scorestruct[0].beatStructure())

qp = ch.asVoice().quantizedPart()
qp.dump()
# qp.dump(tree=False)
```

```python
qp.asScore().show()
```
