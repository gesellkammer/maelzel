```python
from maelzel.core import *
struct = getScoreStruct()
struct.setTempo(60)

ch = Chain("""
    C4-:0.5
    4D+,4Eb:0.5
    4E-:0.5
    4F-:0.5
""")
ch2 = ch.cycle(totaldur=20)
for i, ev in enumerate(ch2):
    if i%3 == 0:
        ev.gliss = True
    if i%5 == 0:
        ev.gliss = False
        ev.setPlay(end=ev.dur*0.25, fade=(0, 0.1))
        ev.addSymbol('staccato')
        prev = ch2.eventBefore(ev.absOffset())
        if prev: 
            prev.addSpanner('slur', ev)
    if i%7 == 2:
        ev.dur *= F(3, 2)
        ch2._update()
    if i%11 == 1:
        ev.dur = ev.dur / 2

ch2.show()
# 
```

```python
ch2.play()
```
