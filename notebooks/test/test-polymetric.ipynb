```python
from maelzel.core import *

struct1 = ScoreStruct("2/4, q=120")
struct2 = ScoreStruct("3/8, q=120")

v1 = Voice([Note(i, 0.5) for i in range(60, 72)], scorestruct=struct1)
v2 = Voice([Note(i, 0.5) for i in range(48, 60)], scorestruct=struct2)

for i in range(v1.numMeasures()):
    n = v1.eventAt((i, 0))
    v1.replaceEvent(n, Chord([n, n + 12], dur=n.dur))

for i in range(v2.numMeasures()):
    n = v2.eventAt((i, 0))
    if n is not None:
        v2.replaceEvent(n, Chord([n, n - 12], dur=n.dur))

    
sco = Score([v1, v2])
sco
```

```python
sco.play(sustain=0.15, kwet=0.85)
```
