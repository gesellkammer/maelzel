## Config

At any moment **there is one (and only one) active Config**. This config encompasses settings related to notation, playback, quantization and many other aspects of *maelzel.core*

```python
c = getConfig()
c
```

The config is at its core a dict where only a well defined set of keys are allowed and values are validated

```python
# Set the default number of channels for playback. Any playback engine created will use this setting
c['play.numChannels'] = 4
```

The next cell will result in an error, since only integers are accepted here:

```python
c['play.numChannels'] = 'four'
```

Customizations can be made persistent by calling `save` on the dictionary (independently of it being active). After that, any config created in a future session will use this dict as its default

```python
c.save()
```

To revert a config to its factory defaults, use `reset`

```python
c.reset()
c['play.numChannels']
```
